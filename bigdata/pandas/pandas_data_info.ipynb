## 판다스 내장 그래프 도구 사용하기

선 그래프

```python
df = pd.read_excel('data/남북한발전전력량.xlsx')
df.head(6)
```

```python
# 남한, 북한 발전전력별 합계 데이터만 추출
df_ns = df.iloc[[0, 5], 2:]
df_ns
```

```python
df_ns.index = ['South', 'North']
df_ns.head()
```

```python
# 선 그래프 그리기
df_ns.plot()
```

```python
# 전치하여 그리기
tdf_ns = df_ns.T
tdf_ns.plot()
```

```python
# 막대 그래프
tdf_ns.plot(kind='bar')
```

```python
# 히스토그램
tdf_ns.plot(kind='hist')
```

```python
# 산점도
df = pd.read_csv('data/auto-mpg.csv', header=None)
df.head()
```

```python
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
df.head()
```

```python
# 2개의 열을 선택하여 산점도 그리기
df.plot(x='weight', y='mpg', kind='scatter')
```

```python
# 박스 플롯
df[['mpg', 'cylinders']].plot(kind='box')
```
