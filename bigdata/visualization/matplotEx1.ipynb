### historam

```python
'''
# hist(bins=)   # bins : 데이터를 집계할 구간 정보
'''
```

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.hist(x, bins=10)
plt.show()
```

```python
# 가로방향
np.random.seed(0)

yLabel = ['A', 'B', 'C', 'D']
yPos = np.arange(4)
yValue = 2+10*np.random.rand(4)

plt.barh(yPos, yValue, alpha=0.5)
plt.yticks(yPos, yLabel)
plt.show()
```

```python
'''
# pie(ratio[비율], explode[간격],labels, colors, autopct[소수점 자리수], shadow, startangle[시작각])
# 원의 형태 유지 : plt.axis('equal')
'''
```

```python
labels = 'A','B','C','D'

ratio = [10,30,40,20]
colors = ['red','skyblue','yellowgreen','pink']
explode = (0,0.1,0,0)
plt.pie(ratio, explode=explode, labels=labels, colors=colors, shadow=True, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

### scatter plot

```python
'''
# 스캐터 플롯(산점도 함수) : 두 개의 실수 데이터 집합의 상관관계를 살펴볼 때 많이 사용하는 차트 
# scatter(x, y)

'''
```

```python
# 정규분포
m, sigma = 0, 0.1
s = np.random.normal(m, sigma, 1000)

plt.hist(s, 30)
plt.grid(True)
plt.show()
```

```python
# scatter
np.random.seed(0)



# np.random.normal(평균, 표준편차, 생픔수)
X = np.random.normal(0, 0.1, 100)
Y = np.random.normal(0, 0.1, 100)
plt.scatter(X,Y)
plt.grid()
plt.show()
```

### bubble chart

```python
'''
# 데이터가 2차원이 아닌 3차원 이상의 경우 점 하나의 크기 또는 컬러를 이용하여 표현
# 크기 : s, 컬러 : c
'''
```

```python
n = 30
np.random.seed(0)
x = np.random.rand(n)
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.pi * (15 * np.random.rand(n))**2

plt.scatter(x, y1, c=y2, s=y3 )
plt.grid()
plt.show()
```
