## 최적화 (Optimization)

```python
# 1차 목적 함수
def f(x):
    return (x-2)**2+2
```

```python
x1 = np.linspace(-1, 4, 100)
x1
```

```python
plt.plot(x1, f(x1))
plt.ylim(0, 10)
plt.grid()
plt.show()
```

```python
# 일단 최적화 값은 2
plt.plot(x1, f(x1))
plt.plot(2, 2, 'ro', ms=10)
plt.ylim(0, 10)
plt.grid()
plt.show()
```

```python
'''
# 목적함수의 값을 가장 최소로 하는 x의 위치를 찾는 것 : 최적화
# 최적화 문제를 해결하는 가장 간단한 방법
 : x값을 여러번 넣어보고 그 중에서 가장 작은 값을 선택하는 방법 (그리드 서치 방법)

# 그리드 서치는 목적함수의 값을 찾기 위한 계산량이 너무 많아지게 됨. 그래서 사용하는 방법이 
  수치적 최적화 (normal optimization)을 사용.

# 수치적 최적화 : 반복적 시행 착오 (trial and error)에 의해 최적화 필요조건을 만족하는 x*를 찾는 방법
               : 함수의 위치가 최저점이 될 때까지 가능한 적은 횟수로 x의 위치를 옮기는 방법
    
# 수치적 최적화를 수행하기 위해 필요한 알고리즘
  1) 어떤 위치 x를 시도한 다음에 다음 번에 시도할 위치 x를 찾는 알고리즘
  2) 현재 위치 x가 최저점인지를 판단할 수 있는 알고리즘 

# 기울기 필요 조건 : 모든 최소점은 기울기가 0이기 때문에 최소점이 되기 위해서는 기울기 값이 0이어야 함.
                    그러나 기울기가 0이어도 최소점이 아닐 수 있다.
# 기울기를 나타내는 함수(벡터)는 g 기호로 표기 : g(gradient)

# 대표적 최적화 알고리즘
  : SGD (Steepest Gradient Descent) 등이 있음
'''
```

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# SGD (최대 경사법) : 현재 위치 x에서의 기울기값 g(x) 만을 이용해서 다음에 시도할 위치를 알아내는 방법
```

```python
# 목적함수
def f1(x):
    return (x-2)**2 + 2

# 도함수
def f1d(x):
    return 2*(x - 2.0)

xx = np.linspace(-1, 6, 100)
plt.plot(xx, f1(xx), 'b-')
plt.plot(xx, f1(xx), 'b-')
```

```python
# step size
mu = 0.4

# 위치 시도값 : k - 보통 0부터 시작
x = 0

xx = np.linspace(-1, 6, 100)
plt.plot(xx, f1(xx), 'b-')

plt.plot(x, f1(x), 'ro', ms=10)
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'g--')

plt.ylim(0, 12)
plt.grid()
plt.show()

print('x = {}, g = {}'.format(x, fid(x)))
```

```python
# 두번째 시도 
x = x - mu * f1d(x)

xx = np.linspace(-1, 6, 100)
plt.plot(xx, f1(xx), 'b-')

plt.plot(x, f1(x), 'ro', ms=10)
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'g--')

plt.ylim(0, 12)
plt.grid()
plt.show()

print('x = {}, g = {}'.format(x, fid(x)))
```

```python
# 세번째 시도 
x = x - mu * f1d(x)

xx = np.linspace(-1, 6, 100)
plt.plot(xx, f1(xx), 'b-')

plt.plot(x, f1(x), 'ro', ms=10)
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'g--')

plt.ylim(0, 12)
plt.grid()
plt.show()

print('x = {}, g = {}'.format(x, fid(x)))
```

```python
# 네번째 시도 
x = x - mu * f1d(x)

xx = np.linspace(-1, 6, 100)
plt.plot(xx, f1(xx), 'b-')

plt.plot(x, f1(x), 'ro', ms=10)
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'g--')

plt.ylim(0, 12)
plt.grid()
plt.show()

print('x = {}, g = {}'.format(x, fid(x)))
```

```python
# 다섯번째 시도 
x = x - mu * f1d(x)

xx = np.linspace(-1, 6, 100)
plt.plot(xx, f1(xx), 'b-')

plt.plot(x, f1(x), 'ro', ms=10)
plt.plot(xx, f1d(x) * (xx - x) + f1(x), 'g--')

plt.ylim(0, 12)
plt.grid()
plt.show()

print('x = {}, g = {}'.format(x, fid(x)))
```

## 보스턴 주택 가격 예측 예제

```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston
```

```python
# 데이터 로딩
boston = load_boston()
boston
```

```python
# 데이터를 dataframe 변환
bostonDf = pd.DataFrame(boston.data, columns=boston.feature_names)
bostonDf.head()
```

* CRIM : 지역별 범죄 발생율

* ZN : 25,000 평방 피트를 초과하는 거주 지역의 비율
    
* INDUS : 비상업지역 넓이 비율
    
* CHAS : 찰스강에 대한 더미 변수 (강의 경계에 위치하면 1, 그렇지 않으면 0)
    
* NOX : 일산화질소 농도

* RM : 거주할 수 있는 방의 개수
    
* AGE : 1940년 이전에 건축된 소유 주택의 비율
    
* DIS : 5개 주요 고용센터까지의 가중 거리
    
* RAD : 고속도로 접근 용이도
    
* TAX : 10,000 달러 당 재산세율

* PTRATIO : 지역의 교사와 학생 수 비율
    
* B : 지역의 흑인 거주 비율
    
* LSTAT : 하위 계층의 비율
    
* MEDV : 본인 소유의 주택 가격 (중앙값)

```python
# 데이터 탐색
bostonDf.shape
```

```python
# label을 dataframe에 추가
bostonDf['PRICE'] = boston.target

bostonDf.shape
```

```python
bostonDf.head()
```

```python
# 컬럼별로 주택 가격에 미치는 영향도를 시각화
fig, axs = plt.subplots(figsize=(16, 8), ncols=4, nrows=2)

lm_features = ['ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'RAD', 'PTRATIO', 'LSTAT']

for i, feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4
    sns.regplot(x=feature, y='PRICE', data=bostonDf, ax=axs[row][col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 레이블 분리
y_target = bostonDf['PRICE']
X_data = bostonDf.drop(['PRICE'], axis=1, inplace=False)

# 학습/테스트 데이터 세트 분리 학습/예측/평가
X_train, X_test, y_train, y_test = train_test_split(X_data, y_target, test_size=0.2, random_state=156)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
print(lr.predict(boston.data[5].reshape(1, -1)))
print('---------------------------------------')
print(boston.target[5])
```

```python
# 모델이 선형회귀식 값
print('절편 값 : ', lr.intercept_)
print('-----------------------------')
print('회귀 계수 : ', np.round(lr.coef_, 1))
```

```python
# 회귀 계수를 컬럼과 같이 확인
coef_boston = pd.Series(data=np.round(lr.coef_, 1), index=X_data.columns)
coef_boston.sort_values(ascending=False)
```

```python
# 평가도구
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print('mse : ', np.round(mse, 3))
print('rmse : ', np.round(rmse, 3))
print('vscore : ', r2_score(y_test, y_pred))
```

```python
'''
# 회귀의 유형
 - 일변량 (Univariate) : 오직 하나의 양적 독립변수 (설명변수)
 - 다변량 (Multivariate) : 두 개 이상의 양적 독립변수(설명변수)
 - 단순 (Simple) : 오직 하나의 종속 변수 (반응변수)
 - 다중 (Multiple) : 두 개 이상의 종속 변수 (반응변수)
 - 선형 (Linear) : 데이터에 대하여 가능한 변환을 취한 후, 모든 계수들이 방정식에 선형적으로 삽입
 - 비선형 (Nonlinear) : 종속변수와 일부 독립변수들의 관계가 비선형이거나 일부 계수들이 비선형적으로 나타남
 - 분산분석 (ANOVA) : 모든 독립변수들이 질적변수들임.
 - 공분산분석 (ANCOVA) : 어떤 독립변수들은 양적변수, 다른 독립변수들은 질적변수
 - 로지스틱 (Logistic) : 종속변수가 질적변수
 
=====================================================================================================
# 대표적인 선형 회귀 모형
 - 일반 선형 회귀
 
 - 릿지 (Ridge) : 선형 회귀에 L2 Regularization 추가한 모형
                : 상대적으로 큰 회귀 계수 값의 예측 영향도를 감소시키기 위한 L2 규제가 적용된 모형
 
 - 라소 (Lasso) : 선형 회귀에 L1 Regularization 추가한 모형
                : 예측 영향이 작은 피처의 회귀 계수를 0으로 만들어 회귀 예측 시 피처가 선택되지 않도록 하는 모형
                : L1 Regularization을 피처 선택 기능으로 불리기도 함
 
 - 엘라스틱넷 (ElasticNet) : L1, L2를 함께 결합한 모형
                          : 피처가 많은 데이터 세트에서 자주 적용 
                          : L1을 통해 피처의 개수를 줄이면서 L2를 통해 계수의 값의 크기를 조정
 
 - 로지스틱 (Logistic) : 상당히 많은 분야에서 분류에 사용되는 선형 모형                          
                      : 일반적으로 이진 분류 뿐만 아니라 희소 영역의 분류(ex. 텍스트 분류와 같은 영역)
                        에서 뛰어난 예측 성능을 보이고 있음
'''
```
